## 载入包

```python
import numpy as np
from sklearn import preprocessing
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import pandas  as pd
%matplotlib inline
import warnings 
warnings.filterwarnings('ignore')
```

## 加载数据-txt文本文件

```python
input_file = 'building_event_binary.txt'
```

```python
X = []
with open(input_file, 'r') as f:
    for line in f.readlines():
        data = line[0:-1].split(',')
        print(data)
        X.append([data[0]] + data[2:])

X = np.array(X)
```

```python
X.shape
```

## 数据清洗

```python
from sklearn import preprocessing
label_encoder = [] # 每一列的转换器
X_encoded = np.empty(X.shape) # 转码后的样本，初始化为空
for i, item in enumerate(X[0]): # 取一行中的特征
    print(i, item)
    if item.isdigit():
        X_encoded[:, i] = X[:, i]
    else:
        le = preprocessing.LabelEncoder()
        le.fit(X[:, i])
        X_encoded[:, i] = le.transform(X[:, i])
        label_encoder.append(le)
X_encoded
X = X_encoded[:, 0:-1].astype(int)
y = X_encoded[:, -1].astype(int)
```

## 模型训练

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
params = {
    'kernel': 'rbf', # 'linear',
    'probability': True,
    'class_weight': 'balanced'
}
```

```python
from sklearn.svm import SVC
clf = SVC(**params)
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

交叉验证

```python
from sklearn.model_selection import cross_val_score

acc = cross_val_score(clf, X, y, scoring='accuracy', cv=5)
acc.mean()
```
