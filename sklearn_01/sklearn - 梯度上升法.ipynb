```python
import numpy as np
import matplotlib.pyplot as plt
```

## 简单抛物线的梯度上升法

```python
def f(x):
    return -(x-50)**2
```

```python
def dx(x):
    return -2*(x-50)
```

```python
x=np.linspace(0,100,20)
```

```python
plt.plot(x,f(x))
plt.show()
```

```python
x_start=90
egt=0.1
compare=1e-8
x_collection=[]
x_last=90
```

```python
while True:
    x_current_dx=dx(x_start)
    x_start=x_start+egt*x_current_dx
    if(abs(x_start-x_last)<compare):
        break
    x_last=x_start
    x_collection.append(x_last)
        
```

```python
x_result=np.array(x_collection)
```

```python
plt.plot(x,f(x))
plt.scatter(x_result,f(x_result),color='red',marker='+')
plt.show()
```

## 正弦函数梯度上升法

```python
x=np.linspace(-10,10,100)
x_start=-3
egt=0.1
compare=1e-8
x_collection=[]
x_last=9

def f(x):
    return np.sin(x)

def dx(x):
    return np.cos(x) 

while True:
    x_current_dx=dx(x_start)
    x_start=x_start+egt*x_current_dx
    if(abs(x_start-x_last)<compare):
        break
    x_last=x_start
    x_collection.append(x_last)
    
x_result=np.array(x_collection)

plt.plot(x,f(x))
plt.show()

plt.plot(x,f(x))
plt.scatter(x_result,f(x_result),color='red',marker='+')
plt.show()
```

## 混合函数梯度上升法

```python
x=np.linspace(-10,10,100)
x_start=-7
egt=0.1
compare=1e-8
x_collection=[]
x_last=9

def f(x):
    return np.sin(x)+np.sin(2*x)

def dx(x):
    return np.cos(x)+2*np.cos(2*x)

plt.plot(x,f(x))
plt.show()

while True:
    x_current_dx=dx(x_start)
    x_start=x_start+egt*x_current_dx
    if(abs(x_start-x_last)<compare):
        break
    x_last=x_start
    x_collection.append(x_last)
    
x_result=np.array(x_collection)

plt.plot(x,f(x))
plt.scatter(x_result,f(x_result),color='red',marker='+')
plt.show()
x_last
```

```python
x_start=10
egt=0.1
compare=1e-8
x_collection=[]
x_last=9

def f(x):
    return np.sin(x)+np.cos(x*1/2)

def dx(x):
    return np.cos(x)-1/2*np.sin(1/2*x)

plt.plot(x,f(x))
plt.show()

while True:
    x_current_dx=dx(x_start)
    x_start=x_start+egt*x_current_dx
    if(abs(x_start-x_last)<compare):
        break
    x_last=x_start
    x_collection.append(x_last)
    
x_result=np.array(x_collection)

plt.plot(x,f(x))
plt.scatter(x_result,f(x_result),color='red',marker='+')
plt.show()
x_last
```
