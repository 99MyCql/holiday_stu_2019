```python
import numpy as np
from matplotlib.pyplot import *
```

## 平面公式

```python
def f(x,y):
    return (x-8)**2 + 2*(y-9)**2

#def f(x,y):
#    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
```

## 分别对x,y求偏导数

```python
def dx(x,y):
    return 2*(x-8)

def dy(x,y):
    return 4*(y-9)
```

## 设置查找的初始位置

```python
x_start=90
y_start=90
```

### 记录最新位置

```python
x_last_dj=5
y_last_dj=5
```

## 设置超参数

```python
compare=1e-8
egt=0.1
```

## 记录数据用于绘图

```python
collection_x=[] 
collection_y=[] 
```

## 开始梯度下降

```python
while True:
    x_current_dx=dx(x_start,y_start)  #对指定x求导数
    x_start=x_start-egt*x_current_dx  #应用梯度计算，x -下降幅度*当前导数
    
    y_current_dy=dy(x_start,y_start)  #对指定x求导数
    y_start=y_start-egt*y_current_dy #应用梯度计算，x -下降幅度*当前导数
    
    if(abs(x_current_dx-x_last_dj)<compare):  #如果上一次和本次导数下降小于定义的精度则表示找到了最小值
        break

    
    if(abs(y_current_dy-y_last_dj)<compare):  #如果上一次和本次导数下降小于定义的精度则表示找到了最小值
        break
        
    collection_x.append(x_start)
    x_last_dj=x_current_dx
    
    collection_y.append(y_start)
    y_last_dj=y_current_dy
    
    
plot(collection_x,collection_y)
scatter(collection_x,collection_y,color='red',marker='+')

n = 256
x = np.linspace(-100,100,n)
y = np.linspace(-100,100,n)
X,Y = np.meshgrid(x,y)
contourf(X, Y, f(X,Y), 20, alpha=0.25, cmap=cm.hot)#alpha=0.25表示颜色 20线的个数
C = contour(X, Y, f(X,Y), 21, colors='black', linewidth=0.01)# 21黑线的个数
clabel(C, inline=0.1, fontsize=10)
show()
```
